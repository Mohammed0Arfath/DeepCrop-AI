```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load your model
disease_model = YOLO("outputs/crop_diseases_aug_model/weights/best.pt", task="segment")

# Run prediction (don't save to disk)
results = disease_model.predict("Crop Disease.png", save=False)

# Get the visualized prediction (as a NumPy image)
img = results[0].plot()

# Show using matplotlib
plt.imshow(img)
plt.title("Predicted Segmentation - Crop Disease")
plt.axis("off")
plt.show()
```
